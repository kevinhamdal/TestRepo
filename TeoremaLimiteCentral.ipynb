```python
from numpy.ma.core import var
import numpy as np
import matplotlib.pyplot as plt


lamda = 0.10  # Parámetro lambda de la distribución exponencial
n = 50  # tamaño de cada muestra individual

numerodemuestras = 10000 #Numero de muestras que se toman
mediasmuestrales = []

media = 1/lamda
var = 1/(lamda**2)

for i in range(numerodemuestras):
  random_numbers = np.random.exponential(scale=media, size=n)
  # ^^^ muestra de n-numeros aleatorias de una distribucion exponencial ^^^
  mediamuestral = np.mean(random_numbers)
  #print(mediamuestral)
  mediasmuestrales.append(mediamuestral)

plt.hist(mediasmuestrales, bins=20, density=True, alpha=0.8)
plt.xlabel("Medias Muestrales")
plt.ylabel("Frecuencia")
plt.title("Histograma")
plt.show()

#Media y varianza Muestrales
mediateo = media
mediaprac = np.mean(mediasmuestrales)
varteo = var/n
varprac = np.var(mediasmuestrales)
print("Media Teorica: ",mediateo)
print("Media Practica: ",mediaprac)
print("Varianza exp: ",varteo)
print("Varianza Practica: ",varprac)
```
